```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pickle
import cv2
import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, Dropout
 
# --- Data Loading and Preprocessing ---
#
# <<< !!! ERROR FIX !!! >>>
# You MUST change this next line. 
# The path below is for KAGGLE, not your computer.
# Replace it with the full path to the "train" folder on your local machine.
#
# Example for Windows: DIRECTORY = r"C:\Users\YourName\Downloads\dogscats\train"
# Example for Mac:     DIRECTORY = "/Users/YourName/Downloads/dogscats/train"
#
DIRECTORY = f"/kaggle/input/dog-vs-cat-fastai/dogscats/train"
CATEGORIES = ['cats', 'dogs']
 
img_size = 100
data = []

print("Starting data loading...")
for category in CATEGORIES: 
    folder = os.path.join(DIRECTORY, category)
    print(f"Loading from: {folder}")
 
    label = 0 if category == "cats" else 1 
 
    for i, img in enumerate(os.listdir(folder)):
        img_path = os.path.join(folder, img)
        img_arr = cv2.imread(img_path)
 
        if img_arr is None: 
            print(f"Warning: could not read image {img_path}")
            continue
 
        img_arr = cv2.resize(img_arr, (img_size, img_size))
        data.append([img_arr, label]) 
 
print("Data loading complete. Shuffling...")
random.shuffle(data)

# --- Prepare data for model ---
X = [features for features, label in data]
y = [label for features, label in data]
 
print(f"Total images loaded: {len(data)}")

# Show a sample image
if len(X) > 0:
    sample_img = X[0]
    print(f"Sample image shape: {sample_img.shape}")
    plt.imshow(sample_img)
    plt.title("Sample Image (Dog or Cat)")
    plt.show()
else:
    print("Error: No data was loaded. Check your DIRECTORY path.")
    # Stop the script if no data is found
    exit()

 
# --- Prepare data for model ---
X = np.array(X) / 255.0
y = np.array(y)
 
# --- Model Building ---
model = Sequential()
model.add(Input(shape=X.shape[1:])) # Shape will be (100, 100, 3)
 
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
 
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
 
model.add(Flatten())
 
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5)) # Dropout for regularization
 
model.add(Dense(2, activation ='softmax'))
 
model.summary()
 
# --- Model Training ---
 
# Compile the model [cite: 398]
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
 
# Train the model [cite: 399]
history = model.fit(X, y, 
                    epochs=10, 
                    validation_split=0.2)
 
# --- Plot Training History ---
print("Plotting training history...")
plt.figure(figsize=(10, 4))
 
# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
 
# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
 
plt.tight_layout()
plt.show()

print("Script finished.")
```
